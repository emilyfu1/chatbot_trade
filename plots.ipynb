```python
import pandas as pd
import matplotlib.pyplot as plt
from statscandatacleaning import processdata
```

```python
alldata = pd.concat([processdata('HS61', 'china'), processdata('HS62', 'china'),
                         processdata('HS61', 'world'), processdata('HS62', 'world'),
                         processdata('HS85', 'china'), processdata('HS85', 'world'),
                         processdata('all', 'china'), processdata('all', 'world')], axis=1)
```

```python
# add 61 and 62
alldata['HS6162china'] = alldata['HS61china'] + alldata['HS62china']
alldata['HS6162world'] = alldata['HS61world'] + alldata['HS62world']

# proportions of china HS6162/85 vs world imports by category
alldata['share_HS6162_byHS2'] = alldata['HS6162china'] / alldata['HS6162world']
alldata['share_HS85_byHS2'] = alldata['HS85china'] / alldata['HS85world']

# proportions of china HS6162/85 vs world total imports
alldata['share_HS6162_worldtotal'] = alldata['HS6162china'] / alldata['allworld']
alldata['share_HS85_worldtotal'] = alldata['HS85china'] / alldata['allworld']

# proportions of china HS6162/85 vs china total imports
alldata['share_HS6162_chinatotal'] = alldata['HS6162china'] / alldata['allchina']
alldata['share_HS85_chinatotal'] = alldata['HS85china'] / alldata['allchina']
```

```python
alldata
```

```python
# HS6162 china / HS 6162 world, HS85 china / HS85 world
alldata[['share_HS6162_byHS2', 'share_HS85_byHS2']].plot()
plt.title('Share of Chinese HS61, HS62 and HS85 imports out of respective HS2 category')
```

```python
# HS6162 china / total world, HS85 china / total world
alldata[['share_HS6162_worldtotal', 'share_HS85_worldtotal']].plot()
plt.title('Share of Chinese HS61, HS62 and HS85 imports out of total imports')
```

```python
# HS6162 china / total china, HS85 china / total world
alldata[['share_HS6162_chinatotal', 'share_HS85_chinatotal']].plot()
plt.title('Share of Chinese HS61, HS62 and HS85 imports out of total imports from China')
```
